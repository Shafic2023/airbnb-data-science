### A/B testing

#### Time for booking

```python
from scipy.stats import norm

mu_A = 300
mu_B = 296

std_A = 105
std_B = 120

n_A = 10000
n_B = 10000

Z = (mu_A - mu_B)/np.sqrt(std_B**2/n_B + std_A**2/n_A)
pvalue = norm.sf(Z)

print("Z-score: {0}\np-value: {1}".format(Z,pvalue))
```

```python
z = np.arange(-3, 3, 0.1)
fig = plt.figure()
plt.plot(z, norm.pdf(z))
#plt.plot(z, norm.sf(z))
plt.fill_between(z[z>Z], norm.pdf(z[z>Z]), color='k')
plt.text(0.001, 0.75, "Z-score: " + str(np.round(Z,2)), transform = ax.transAxes, fontsize=12, fontweight='bold')
plt.text(0.001, 0.65, "p-value: " + str(np.round(pvalue,4)), transform = ax.transAxes, fontsize=12, fontweight='bold')
plt.xlabel("z-score")
plt.legend(["pdf", "p-value"])
plt.title('Time to Complete Booking: z-test for Control vs Variation', fontsize=12)
plt.savefig("hypothesis1_pdf_pvalue.jpg", facecolor='w', transparent=True)
plt.show()
```

#### Conversion Rate

```python
from scipy.stats import chi2
T = np.array([165, 165, 9835, 9835])
# O = np.array([90, 80, 5910, 3920])
O = np.array([150, 180, 9850, 9800])

D = np.sum(np.square(T-O)/T)

pvalue = chi2.sf(D, df=1)

print("distance d: {0}\np-value: {1}".format(D,pvalue))

d = np.arange(0, 5, 0.1)
plt.plot(d, chi2.pdf(d, df=1))
plt.plot(d, chi2.sf(d, df=1))
plt.fill_between(d[d>D], chi2.pdf(d[d>D], df=1))
plt.text(0.001, 0.75, "Z-score: " + str(np.round(D, 2)), transform = ax.transAxes, fontsize=12, fontweight='bold')
plt.text(0.001, 0.65, "p-value: " + str(np.round(pvalue,5)), transform = ax.transAxes, fontsize=12, fontweight='bold')
plt.xlabel("Squared relative distance between the theoretical and the observed distributions", fontsize=11)
plt.legend(["pdf", "p-value"])
#plt.title(r'$\chi^2$' + distribution of Conversion Rate Values)
plt.title("Conversion Rate " + r'$\chi^2$' + " Distribution" + ":  Control vs Variation", fontsize=12)
plt.savefig("hypothesis2_pdf_pvalue.jpg")
plt.show()
```
